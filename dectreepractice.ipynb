```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.read_csv("dataset.csv")
label_encoder = LabelEncoder()
encoded_df = df.apply(label_encoder.fit_transform)
```

```python
X = encoded_df.drop("Destination", axis=1)
y = encoded_df["Destination"]
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(X, y)
```

```python
class_names = label_encoder.inverse_transform(model.classes_)
```

```python
tree.plot_tree(model, feature_names=X.columns, class_names = class_names, filled=True, fontsize=12)
```
